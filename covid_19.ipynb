```python
import pandas as pd
```

```python
url_confirmed_US= "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/adf97ebf575dd32238251adb809e671c8810b213/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
```

```python
confirmed_US= pd.read_csv(url_confirmed_US)
```

```python
confirmed_US= confirmed_US[["Province_State", "Country_Region","Lat", "Long_", "7/20/20"]]
```

```python
confirmed_US.rename(columns={"7/20/20": "Confirmed"}, inplace=True)
```

```python
confirmed_US.rename(columns={"Province_State": "Province/State"}, inplace= True)
confirmed_US.rename(columns={"Country_Region": "Country/Region"}, inplace= True)
confirmed_US.rename(columns={"Long_": "Long"}, inplace= True)
```

```python
confirmed_US["Date"]= "7/20/20"
```

```python
confirmed_US.head()
```

```python
url_confirmed_global= "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/adf97ebf575dd32238251adb809e671c8810b213/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```python
confirmed_global= pd.read_csv(url_confirmed_global)
```

```python
confirmed_global=confirmed_global[["Province/State", "Country/Region", "Long", "Lat","7/20/20"]]
```

```python
confirmed_global.rename(columns={"7/20/20": "Confirmed"}, inplace= True)
```

```python
confirmed_global["Date"]= "7/20/20"
```

```python
confirmed_global.head()
```

```python
url_death_US ="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/adf97ebf575dd32238251adb809e671c8810b213/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv"
```

```python
death_us= pd.read_csv(url_death_US)
```

```python
death_us= death_us[["Province_State","Country_Region", "Lat", "Long_",'7/20/20' ]]
```

```python
death_us.rename(columns={"Province_State": "Province/State"}, inplace= True)
death_us.rename(columns={"Country_Region": "Country/Region"}, inplace= True)
death_us.rename(columns={"Long_": "Long"}, inplace= True)
death_us.rename(columns={"7/20/20": "death"}, inplace= True)
```

```python
death_us["Date"]= "7/20/20"
```

```python
death_us.head()
```

```python
url_death_global= "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/adf97ebf575dd32238251adb809e671c8810b213/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```python
death_global= pd.read_csv(url_death_global)
```

```python
death_global=death_global[["Province/State", "Country/Region", "Long", "Lat","7/20/20"]]
death_global.rename(columns={"7/20/20": "death"}, inplace= True)
```

```python
death_global["Date"]= "7/20/20"
```

```python
death_global.head()
```

```python
url_recovered_global= "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/adf97ebf575dd32238251adb809e671c8810b213/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
```

```python
recovered_global= pd.read_csv(url_recovered_global)
```

```python
recovered_global= recovered_global[['Province/State', 'Country/Region', 'Lat', 'Long', '7/20/20' ]]
recovered_global["Date"]='7/20/20'
```

```python
recovered_global.rename(columns={"7/20/20": "recovered"}, inplace= True)
```

```python
# combaining US confirmed data with global confirmed data
combined_confirmed= pd.concat([confirmed_global, confirmed_US], axis=0)
```

```python
combined_confirmed.tail()
```

```python
# combaining US death data with global death data
combined_death= pd.concat([death_global, death_us], axis=0)
```

```python
combined_death.tail()
```

```python
# Merging multiple data using inner 


confirmed_death_merge= pd.merge(combined_confirmed, combined_death,  how= "right", on= ["Province/State", "Country/Region","Long","Lat","Date"])
```

```python
result= pd.merge(confirmed_death_merge, recovered_global,  how= "right", on= ["Province/State", "Country/Region","Long","Lat","Date"])
```
